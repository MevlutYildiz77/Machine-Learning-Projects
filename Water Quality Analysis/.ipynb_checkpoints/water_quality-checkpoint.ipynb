```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WaterQuality.csv", sep= ";")
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
import plotly.express as px
plt.figure(figsize=(12,10))
sns.countplot(df["Potability"])
plt.title("Distribution of Potability or Unfit")
plt.show()
```

```python
import plotly.express as px
data = df
figure = px.histogram(df, x= "ph", color= "Potability",
                      title="Ph Factor to Potability")
figure.show()
```
