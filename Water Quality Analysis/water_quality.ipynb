```python
df = pd.read_csv("WaterQuality.csv", sep= ";")
df.head()
```

```python
df.info()
```

```python
df["ph"] = df["ph"].str.strip(".")
df.head()
```

```python
df["ph"] = df['ph'].astype(float)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
import plotly.express as px
plt.figure(figsize=(15,10))
sns.countplot(df.Potability)
plt.title("Distribution of Potability or Unfit")
plt.show()
```
